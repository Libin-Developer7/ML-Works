```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
sw = stopwords.words("english")
print(sw)
```

```python
sentence = "india won their third t20 world cup. Sanju samson rewarded with player of the tournament"
new_sentence = [w for w in sentence.split(" ") if w not in sw]
new_sentence = " ".join(new_sentence)
print(new_sentence)
```

```python
import re
cleaned_data = re.sub("[^a-zA-Z0-9 ]","",new_sentence)
cleaned_data
```

```python
text = "the sun rises in the east"
nltk.download("punkt_tab")
tokens = nltk.word_tokenize(text)
print(tokens)
filtered_tokens = [t for t in tokens if t not in sw]
print(filtered_tokens)
```

```python
sentence = "the# jupyter$ is@ the second@ largest@@ planet"
# remove special characters
cleaned_text = re.sub("[^a-zA-Z ]","",sentence)
print(cleaned_text)
```

```python
# tokenize
tokens = nltk.word_tokenize(cleaned_text)
print(tokens)
```

```python
# remove stopwords
filtered_tokens = [t for t in tokens if t not in sw]
print(filtered_tokens)
filtered_sentence = " ".join(filtered_tokens)
print(filtered_sentence)
```

```python
#ntlk>stem>PorterStemmer,SnowballStemmer
from nltk.stem import PorterStemmer
ps =PorterStemmer()
for t in tokens:
    print(f"{t[0]}:{ps.stem(t[0])}")
```

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer(language="english")
for t in tokens:
    print(f"{t[0]}:{ss.stem(t[0])}")
```
