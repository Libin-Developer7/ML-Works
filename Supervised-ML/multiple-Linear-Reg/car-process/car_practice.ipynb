```python
df["brand"] = df["Name"].str.strip().str.split(" ").str[0]
print(df["brand"])
df.drop(columns="Name",inplace=True)
```

```python
df["mileage"] = df["mileage"].str.split(" ").str[0]
df["mileage"] = df["mileage"].astype(float)
print(df["mileage"].dtype)
print(df["mileage"])
print(df["mileage"].isnull().sum())
```

```python
df["engine"] = df["engine"].str.split("C").str[0]
df["engine"] = df["engine"].astype(float)
df["engine"]
```

```python
df["max_power"] = df["max_power"].str.split("b").str[0].str.strip()
df[df["max_power"] == ""]
df["max_power"] = df["max_power"].replace("",np.nan)
df["max_power"] = df["max_power"].astype(float)
```

```python
df["torque"] = df["torque"].str.split("N|n|@|k|\(|K|/").str[0]
df["torque"] = df["torque"].astype(float)
df["torque"].dtype
```

```python
df.columns[7:12]
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy="most_frequent")
df.iloc[:,7:12] = imputer.fit_transform(df.iloc[:,7:12])
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,:]
x.drop(columns="selling_price",inplace=True)
x.head()
```

```python
y = df.iloc[:,1]
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([("encoder",OneHotEncoder(sparse_output=False),["brand","fuel","seller_type","transmission","owner"])],remainder="passthrough")
x = pd.DataFrame(ct.fit_transform(x),columns=ct.get_feature_names_out())
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred)*100)
```
